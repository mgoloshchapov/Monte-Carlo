```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
from numba import njit
```

```python
@njit
def rnd(x, del_a):  # Trial standart devitations
    # del_a: Trial standard deviation
    xx = np.random.normal(x, del_a)
    return xx


@njit
def density(x):  # Probability density Cauchy
    gam = 1
    x0 = 30
    return 1 / np.pi * (gam / ((x - x0) ** 2 + gam ** 2))


@njit
def metropolis_hastings(target_density, x0, del_a, size=50000):
    burnin_size = 10000
    size += burnin_size
    xt = x0  # The beginning of the Markov chain
    samples = []
    k = 0
    acc = []
    for i in range(size):
        xt_candidate = rnd(xt, del_a)
        accept_prob = (target_density(xt_candidate))/(target_density(xt))
        if np.random.uniform(0, 1) < accept_prob:
            k += 1
            xt = xt_candidate
        samples.append(xt)
        acc.append(accept_prob)
    print('Accept', k/size)  # Probability of acceptance
    samples = np.array(samples[burnin_size::2])  # Prorezhivanie vyborki
    return np.transpose(samples)
```

```python
from scipy.special import voigt_profile


def rnd(x, del_a):  # Trial standart devitations
    # del_a: Trial standard deviation
    xx = np.random.normal(x, del_a)
    return xx



def density(x):  # Probability density Cauchy
    sigma = 1
    gamma = 0.0
    return voigt_profile(x, sigma, gamma)



def metropolis_hastings(target_density, x0, del_a, size=50000):
    burnin_size = 10000
    size += burnin_size
    xt = x0  # The beginning of the Markov chain
    samples = []
    k = 0
    acc = []
    for i in range(size):
        xt_candidate = rnd(xt, del_a)
        accept_prob = (target_density(xt_candidate))/(target_density(xt))
        if np.random.uniform(0, 1) < accept_prob:
            k += 1
            xt = xt_candidate
        samples.append(xt)
        acc.append(accept_prob)
    print('Accept', k/size)  # Probability of acceptance
    samples = np.array(samples[burnin_size::2])  # Prorezhivanie vyborki
    return np.transpose(samples)
```

```python
from tqdm import tqdm
def Gelman_Rubin(del_a):
    a = []
    J = 10
    for i in range(J):
        a.append(metropolis_hastings(density, np.random.normal(0, 1e-3), del_a))
    a = np.asarray(a)
    L = len(a[0])
    x_av = []
    for i in range(J):
        x_av.append(np.average(a[i]))
    x_star = np.average(x_av)
    x_av = np.asarray(x_av)
    B = L/(J-1) * sum((x_av-x_star)**2)
    W = 0
    for j in range(J):
        W += 1/(L-1) * sum((a[j]-x_av[j])**2) 
    W = 1/J * W
    print("\nDispersion =", B)
    R = ((L-1)/L * W - 1/L * B)/W
    print("\nGelman-Rubin statistic =", R)
    return R

b = []
de = np.linspace(1e-3, 1e3, 10)
for i in tqdm(de):
    b.append(Gelman_Rubin(i))
plt.plot(de, b)
plt.show()
    
```
