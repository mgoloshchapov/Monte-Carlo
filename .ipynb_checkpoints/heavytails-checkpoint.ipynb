### Metropolis-Hastings sampling from Voigt distribution

```python
def sampler(p, N, step=1e-1):
    q = norm(0.0, step)
    delta = q.rvs(N-1)
    u = np.random.uniform(0, 1, N-1)
    accepted = 0

    xc = np.random.normal(0.0, 1.0)
    samples = [xc]

    for i in range(N-1):
        xt = xc + delta[i]
        a = p(xt) * q.pdf(xc-xt) / (p(xc) * q.pdf(xt-xc))
        
        if a > u[i]:
            xc = xt
            accepted += 1

        samples.append(xc)

    p_accept = accepted / (N-1)

    return np.array(samples), p_accept
```

```python
step = 1e-1
N = int(1e5)
q = norm(0.0, step)
delta = q.rvs(N-1)

plt.hist(delta, bins=100)
```

```python
# p = lambda x: voigt_profile(x, 1.0, 0.1)

def voigt(x, alpha):
    return voigt_profile(x, alpha, 1.0-alpha)

p = lambda x: voigt(x, 0.5)

samples, acc_rate = sampler(p, 2000, step=1e0)
print(acc_rate)

x = np.linspace(-5,5,100)
plt.hist(samples, density=True, bins=50)
plt.plot(x, p(x))
```

```python
N = int(1e3)

alpha = np.linspace(0, 1, 21)
mean = []
std = []
acceptance = []

for a in alpha:
    p = lambda x: voigt(x, a)
    samples, acc_rate = sampler(p, 2000, step=1e0)

    mean_t = np.mean(samples)
    std_t = np.sqrt(np.mean(samples ** 2) - mean_t**2)

    std.append(std_t)
    mean.append(mean_t)
    acceptance.append(acc_rate)
```

```python
plt.plot(alpha, mean, label="mean")
plt.plot(alpha, std, label="std")

plt.xlabel("$\\alpha$", fontsize=14)
plt.title("$\sigma = \\alpha, \; \gamma = 1-\\alpha, \; \; N=1e4$")
plt.legend()
```

### Comparison of Metropolis-Hastings with different hyperparameters

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t as student
from scipy.stats import norm, cauchy
from scipy.special import rel_entr

from numba import jit
```

```python
@jit
def sampler(p, q, N, burnout=0):
    delta = q.rvs(N-1 + burnout)
    u = np.random.uniform(0, 1, N-1 + burnout)
    accepted = 0

    xc = np.random.normal(0.0, 1.0)
    samples = [xc]

    for i in range(N-1 + burnout):
        xt = xc + delta[i]
        a = p(xt) * q.pdf(xc-xt) / (p(xc) * q.pdf(xt-xc))
        
        if a > u[i]:
            xc = xt
            accepted += 1

        samples.append(xc)

    p_accept = accepted / (N-1 + burnout)

    return np.array(samples[burnout:]), p_accept


def KL_div(p, samples, bins):
    q_probs, _ = np.histogram(samples, bins=bins, density=True)
    p_probs = p(bins[1:])
    bin_step = bins[1] - bins[0]
    
    return np.sum(rel_entr(q_probs, p_probs)) * bin_step
```

```python
q = student(4)
gauss = norm()
lorentz = cauchy()

x = np.linspace(-5,5,100)

plt.plot(x, q.pdf(x))
plt.plot(x, gauss.pdf(x))
plt.plot(x, lorentz.pdf(x))
```

```python
x0 = 1.5
p = lambda x: (np.exp(-(x-x0)**2) + np.exp(-(x+x0)**2)) / (2* np.sqrt(np.pi))
q = student(df=5, scale=0.5)

bins = np.linspace(-5, 5, 50)

samples, p_accept = sampler(p, q, N=int(1e3), burnout=int(1e3))
plt.hist(samples, bins=bins, density=True)
plt.plot(bins, p(bins))

print(KL_div(p, samples, bins))
```

```python
scales = np.logspace(-3,1,20)
res = []

for s in scales:
    q = student(df=5, scale=s)
    samples, p_acc = sampler(p, q, N=int(1e3), burnout=int(1e3))
    res.append(KL_div(p, samples, bins))
```

```python
def Gelman_Rubin(p, q, n_samples, n_chains, burnout=0):
    means = []
    vars = []
    for i in range(n_chains):
        samples, _ = sampler(p, q, N=n_samples, burnout=burnout)
        means.append(np.mean(samples))
        vars.append(np.var(samples, ddof=1))

    B = np.var(means, ddof=1) * n_samples
    W = np.mean(vars)

    R = ((n_samples - 1)/n_samples * W + 1/n_samples * B)/W

    return R
```

```python
q = student(df=opt_df, scale=opt_scale)

sample_sizes = np.arange(int(2e2), int(10e3), int(4e2))
Gelman_Rubin_res = []

for n_samples in sample_sizes:
    Gelman_Rubin_res.append(Gelman_Rubin(p, q, n_samples, n_chains=10))
```
